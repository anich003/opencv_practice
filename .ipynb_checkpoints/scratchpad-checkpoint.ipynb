```python
def generateMask(img,cpy):
    """
    Takes original and annotated image and returns mask,
    where each color is a different label
    
    Input:
    img    -    original image
    cpy    -    annotated image
    
    Returns:
    mask   -    Blacked out image, except for annotations
    """
    
    xs,ys = np.where(cpy!=img)[:-1]
    mask  = np.zeros_like(cpy)
    mask[xs,ys] = cpy[xs,ys]
    return mask
```

# Statistics

```python
from scipy import linalg

def mvn_likelihood(x,mu,cov):
    k = len(mu)
    A = np.log(linalg.det(cov))
    B = (x-mu).dot(linalg.inv(cov).dot((x-mu)))
    C = k*np.log(2*np.pi)
    return np.exp(-0.5 * (A + B + C))
```

### Collect labeled pixels and calculate statistics

```python
mus    = []
covs   = []
priors = []

for label,color in label_colors.items():
    
    xs, ys = np.where(np.all(mask==color, axis=-1))
    if (len(xs) > 0) and (len(ys) > 0):
        pixels = img[xs,ys]

        mus.append(pixels.mean(axis=0))
        covs.append(np.cov(pixels.T))
        priors.append(pixels.shape[0])

priors = np.array(priors)/sum(priors)

mu1,mu2 = mus
cov1,cov2 = covs
```

```python
like_1 = np.apply_along_axis(mvn_likelihood,2,img,mu1,cov1)
like_2 = np.apply_along_axis(mvn_likelihood,2,img,mu2,cov2)

den = like_1*priors[0]+like_2*priors[1]
num = like_1*priors[0]
posterior = num/den
```

```python
fig,axes = plt.subplots(2,2,figsize=(10,8))

ax0,ax1,ax2,ax3 = axes.ravel()

ax0.imshow(cpy)
ax1.imshow(like_1, cmap=plt.cm.afmhot)
ax2.imshow(like_2, cmap=plt.cm.afmhot)
ax3.imshow(posterior>0.5, cmap=plt.cm.gray)

for ax in axes.ravel():
    ax.axis('off')
    
#fig.tight_layout()
fig.subplots_adjust(wspace=0.1,hspace=0.001)
```
